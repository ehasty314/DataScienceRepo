```python
import pandas as pd

# Read the data
df = pd.read_csv('/Users/ehast/Downloads/train.csv')

# Check missing values for each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Mileage'].info()
```

```python
df['Mileage'].isna().sum()
```

```python
df = df.dropna(subset=['Mileage']) # Dropping as only two values missing is enough to safely drop without
# statistically significant loss
```

```python
df['Mileage'].isna().sum()
```

```python
df['Engine'].info() # Let's impute these, as a mean will be valuable
```

```python
df['Mileage'] = df['Mileage'].str.split(' ', expand=True)[0].astype(float)
df['Engine'] = df['Engine'].str.split(' ', expand=True)[0].astype(float)
df['Power'] = df['Power'].str.split(' ', expand=True)[0].astype(float)
df['New_Price'] = df['New_Price'].str.split(' ', expand=True)[0].astype(float)
```

```python
df['Engine'].fillna(df['Engine'].mean(), inplace=True)
```

```python
df['Power'].unique() # Let's median this one, as there are some high values that may skew the mean
```

```python
df['Power'].fillna(df['Power'].median(), inplace=True)
```

```python
df['New_Price'].unique() # Again, median will be best due to skewed values from range of numbers
```

```python
df['New_Price'].fillna(df['New_Price'].median(), inplace=True)
```

```python
df['Seats'].unique() # Mode will work best for this table
```

```python
mode_value = df['Seats'].mode()[0]
df['Seats'].fillna(mode_value, inplace=True)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Convert 'Fuel_Type' and 'Transmission' into one hot encoded values
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], drop_first=True)
```

```python
current_year = 2023
df['Car_age'] = current_year - df['Year']
```
