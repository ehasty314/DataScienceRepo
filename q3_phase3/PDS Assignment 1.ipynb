```python
import pandas as pd
# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/ehast/Downloads/StudentsPerformance.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
print(df.isnull().sum().sum())
```

```python
#Bar charts on test scores by gender:
import matplotlib.pyplot as plt

df.groupby('gender').mean()[['math score', 'reading score', 'writing score']].plot(kind='bar')
plt.title("Average Scores by Gender")
plt.ylabel("Score")
plt.show()

#This makes comparing the average performances across different categories easier.
```

```python
#Boxplost for distribution analysis:
df.boxplot(column='math score', by='gender')
plt.title("Math Score Distribution by Gender")
plt.suptitle("")  # Suppress default title
plt.ylabel("Score")
plt.show()

#This makes understanding the distribution of scores for different groups, 
#including median scores, quartiles, and potential outliers easier.
```

```python
#Histograms for score distribution:
df['math score'].hist(bins=20)
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()

# This makes grasping the overall distribution of scores for the entire cohort easier.
```

```python
#Pie chart for data distribution
df['race/ethnicity'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribution of Students by Race/Ethnicity")
plt.ylabel("")  # Remove the default 'race/ethnicity' ylabel
plt.show()

#This makes observing the distribution of categorical variables easier.
```

```python
#Scatter plots to identify correlation:
df.plot.scatter(x='math score', y='reading score')
plt.title("Scatter Plot between Math and Reading Scores")
plt.show()

#This makes identifying relationships and potential correlations between two continuous variables easier.
```
